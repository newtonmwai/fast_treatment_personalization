#### Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split

Categorical_cols = {'PTGENDER': 2, 'PTEDUCAT': 13, 'PTETHCAT': 3, 'PTRACCAT': 7, 'PTMARRY': 5, 'APOE4': 3, 'DX': 3, 'PTGENDER_prev': 2,
                    'PTEDUCAT_prev': 13, 'PTETHCAT_prev': 3, 'PTRACCAT_prev': 7, 'PTMARRY_prev': 5, 'APOE4_prev': 3, 'prev_DX': 3, 'A_Cat_prev': 8}

#grouped_id_data = data.groupby('RID')
#datanew = pd.DataFrame({})
#for i in range(sample_size):
#    df = grouped_id_data.get_group(i)
#    df['prev_DX'] = df['DX'].shift(periods=1).fillna(-1)
#    df['prev_Y_hat'] = df['Y_hat'].shift(periods=1).fillna(170)
#    datanew = datanew.append(df, ignore_index=True)
    
def check_categorical(df, cols, Y_col, categorical_cols=Categorical_cols):
    for col in cols:
        #print("\ncheck_categorical: ", col)
        # OR any(s in col for s in categorical_cols)
        if(((col in categorical_cols)) and (col != Y_col)):
            #print("\ncheck_categorical,  True: ", col)
            df = df.astype({col: int})
            # print(df.head())
            df = pd.concat([df, pd.get_dummies(df[col].astype(pd.CategoricalDtype(
                categories=list(range(categorical_cols[col])))), prefix=col, drop_first=True)], axis=1)
            #df = pd.concat([df, pd.get_dummies(df[col], prefix=col, drop_first=True)], axis=1)
            df.drop([col], axis=1, inplace=True)
    return df

def pre_processing(df1):
    # Columns to be standardized
    standardize_x_cols = ['TAU', 'PTAU', 'FDG', 'AV45', 'prev_Y_hat']

    df = df1.copy()
    
    sc = StandardScaler()
    df[standardize_x_cols] = sc.fit_transform(df[standardize_x_cols])
    # creating instance of labelencoder
    labelencoder = LabelEncoder()
    df['A_Cat'] = labelencoder.fit_transform(df['A'])
    df.drop('A', axis=1, inplace=True)
    
    df = check_categorical(df, X_cols, 'Y_hat')    
    return df
```

#### Simple model that predicts $reward$ given $Context$

```python
class SimpleModel2:
    def __init__(self, e, model, col='ADAS13'):
        self.model = model
        self.e = e
        self.col = col
    
    def predict(self, x, z):
        return self.gen_potential_Outcomes(x, z, self.e.gamma, self.e.gamma, self.e.P1, col=self.col)
    
    def gen_potential_Outcomes(self, x, z, gamma_0, gamma_1, P1, col='ADAS13'):
        Y_s = {}
        X_cols = ['prev_ADAS13', 'PTETHCAT', 'PTRACCAT', 'PTEDUCAT',
                     'PTGENDER', 'PTMARRY', 'TAU', 'PTAU', 'APOE4', 'FDG', 'AV45',  'ABETARatio'] # 'Z'
        
        x = x[X_cols]
        x['Z'] = z
        #z = x['Z']
        
        noise = ar.sample_asymetric(residuals[self.col].values, len(x))
        ADAS13 = self.model.predict(dm.check_categorical(x, list(x.columns), '_')) + noise
        
        Y_s = np.array([tr.assign_treatment_effect(a, z, gamma_0, gamma_1, P1) for a in range(8)]) + ADAS13 
        return -(Y_s - Y_s[0])
```

```python
# 
A = 8
dimension = 33
cs = e.reset()
X_cols = ['prev_ADAS13', 'PTETHCAT', 'PTRACCAT', 'PTEDUCAT',
                     'PTGENDER', 'PTMARRY', 'TAU', 'PTAU', 'APOE4', 'FDG', 'AV45',  'ABETARatio']

x = e.buffer_[X_cols]
x = np.array(dm.check_categorical(x, list(x.columns), '_'))

beliefs = [LinearGaussianBelief(d=dimension, v=1) for i in range(A)]
context_ts = ThompsonSampling(beliefs=beliefs)

action, _, info = context_ts.act(x.reshape(dimension,))
_, r, _, info = e.step(action)
print(f'Action {action} reward {r}')
context_ts.update(x, action, r)
```

```python
sample_size = 1000
e = ADCBEnvironment(gamma=2, 
                    epsilon=0.1, 
                    policy='DX_Based',
                    regenerate=False,
                    horizon=6,
                    n_buffer=sample_size,
                    rnd_seed=1)

print('Time spent fitting env: %.2fs' %(e.fit_time))
cs = e.reset()

N_samples = 3
T = 100
A = 8
dimension = 33
rfr = e.load_cov_model('ADAS13', autoreg=True)
```

```python
N_rewards = {}
N_regrets = {}

for p in range(N_samples):
    cs = e.reset()
    regrets = np.zeros(T)

    x = e.buffer_[X_cols]
    x = np.array(dm.check_categorical(x, list(x.columns), '_'))
    x = x.reshape(dimension,)

    beliefs = [LinearGaussianBelief(d=dimension) for i in range(A)]
    context_ts = ThompsonSampling(beliefs=beliefs)

    for t in range(T):
        action, _, info = context_ts.act(x)
        _, r, _, info = e.step(action)
        context_ts.update(x, action, r)
        regrets[t] = info['regret']

    N_rewards[p] = info['expected_rewards']
    N_regrets[p] = regrets

N_rewards_df = pd.DataFrame(N_rewards)
rewards_mean = list(N_rewards_df.mean(axis=1))

N_regrets_df = pd.DataFrame(N_regrets)
regrets_mean = list(N_regrets_df.mean(axis=1))

### Latent Contextual TS (mTS)
N_rewards_model = {}
N_regrets_model = {}


for p in range(N_samples):
    cs = e.reset()

    x = e.buffer_
    
    regrets = np.zeros(T)
    
    model = SimpleModel2(e, rfr)
    #rmses= np.array([[1] * 8, [1] * 8])
    
    belief = LatentModelBelief(model, random_state=p) #prior=.43)# , 
    model_ts = ModelThompsonSampling(model=model, belief=belief)

    for t in range(T):
        action, _, info_act = model_ts.act(x)
        _, r, _, info = e.step(action)
        model_ts.update(x, info_act['z'], action, r)
        regrets[t] = info['regret']

    N_rewards_model[p] = info['expected_rewards']
    N_regrets_model[p] = regrets
    
N_rewards_model_df = pd.DataFrame(N_rewards_model)
rewards_model_mean = list(N_rewards_model_df.mean(axis=1))

N_regrets_model_df = pd.DataFrame(N_regrets_model)
regrets_model_mean = list(N_regrets_model_df.mean(axis=1))
```

```python
rewards_std = list(N_rewards_df.std(axis=1))
regrets_std = list(N_regrets_df.std(axis=1))

rewards_model_std = list(N_rewards_model_df.std(axis=1))
regrets_model_std = list(N_regrets_model_df.std(axis=1))

plt.figure()
plt.plot(rewards_model_mean)
plt.plot(rewards_mean, linestyle='dotted')
plt.grid(zorder=-100)
plt.ylabel('Expected reward')
plt.xlabel('t')
plt.legend([r'mTS Contextual $\sigma$=1', 'Linear Contextual TS'])

plt.figure()
plt.plot(np.cumsum(regrets_model_mean))
plt.plot(np.cumsum(regrets_mean), linestyle='dotted')
plt.ylabel('Cumulative regret')
plt.xlabel('t')
plt.grid(zorder=-100)
plt.legend([r'mTS Contextual $\sigma$=1', 'Linear Contextual TS'])

#plt.errorbar(list(range(len(rewards_model_mean))), rewards_model_mean, yerr=rewards_model_std)
#plt.errorbar(list(range(len(rewards_mean))), rewards_mean, yerr=rewards_std, linestyle='dotted')
```

```python
sample_size = 1000
e = ADCBEnvironment(gamma=2, 
                    epsilon=0.1, 
                    policy='DX_Based',
                    regenerate=False,
                    horizon=6,
                    n_buffer=sample_size,
                    rnd_seed=1)

print('Time spent fitting env: %.2fs' %(e.fit_time))
cs = e.reset()

N_rewards = {}
N_regrets = {}

for p in range(N_samples):
    cs = e.reset()
    regrets = np.zeros(T)

    x = e.buffer_[X_cols]
    x = np.array(dm.check_categorical(x, list(x.columns), '_'))
    x = x.reshape(dimension,)

    beliefs = [LinearGaussianBelief(d=dimension) for i in range(A)]
    context_ts = ThompsonSampling(beliefs=beliefs)

    for t in range(T):
        action, _, info = context_ts.act(x)
        _, r, _, info = e.step(action)
        context_ts.update(x, action, r)
        regrets[t]=info['regret']

    N_rewards[p] = info['expected_rewards']
    N_regrets[p] = regrets

N_rewards_df = pd.DataFrame(N_rewards)
rewards_mean = list(N_rewards_df.mean(axis=1))

N_regrets_df = pd.DataFrame(N_regrets)
regrets_mean = list(N_regrets_df.mean(axis=1))

N_rewards_model = {}
N_regrets_model = {}


for p in range(N_samples):
    cs = e.reset()

    x = e.buffer_
    
    regrets = np.zeros(T)
    
    model = SimpleModel2(e, rfr)
    #rmses= np.array([[1] * 8, [1] * 8])
    
    belief = LatentModelBelief(model, random_state=p, sigma=2) #prior=.43, , 
    model_ts = ModelThompsonSampling(model=model, belief=belief)

    for t in range(T):
        action, _, info_act = model_ts.act(x)
        _, r, _, info = e.step(action)
        model_ts.update(x, info_act['z'], action, r)
        regrets[t] = info['regret']

    N_rewards_model[p] = info['expected_rewards']
    N_regrets_model[p] = regrets
    
N_rewards_model_df = pd.DataFrame(N_rewards_model)
rewards_model_mean = list(N_rewards_model_df.mean(axis=1))

N_regrets_model_df = pd.DataFrame(N_regrets_model)
regrets_model_mean = list(N_regrets_model_df.mean(axis=1))
```
