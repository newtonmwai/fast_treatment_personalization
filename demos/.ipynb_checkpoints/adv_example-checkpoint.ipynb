```python
%load_ext autoreload
%autoreload 2

import warnings
warnings.filterwarnings("ignore")

import os, sys, pickle
import seaborn as sns

sys.path.append('../')

from healthy_gym.agents.pure_explorer import *
from healthy_gym.agents.bayes_design import *
from healthy_gym.agents.beliefs import *
from healthy_gym.agents.lp_explorer import *



from healthy_gym.environments.gaussian_environment import GaussianEnvironment

from healthy_gym.experiment import PureExploreExperiment

import matplotlib.pyplot as plt


%matplotlib inline

plt.rc('font', size=24, family='serif')
```

Hard Example

```python
means = np.array([[1, 0.5, 0.9], [0.9, 1, 0.6], [0.7, 0.9, 1], [1.5, 1.1, 0.5]])
plt.scatter([0, 1, 2], means[0, :], label=r'$Z_0$') 
plt.scatter([0, 1, 2], means[1, :], marker='*', label=r'$Z_1$')
plt.scatter([0, 1, 2], means[2, :], marker='^', label=r'$Z_2$')
plt.scatter([0, 1, 2], means[3, :], marker='+', label=r'$Z_3$')
plt.ylabel(r'$\mu_a$')
plt.xlabel('Arm')
plt.legend(bbox_to_anchor=(1.1, 1.05))
```

```python
env = GaussianEnvironment(means, std=1)
env.reset()
models = env.get_models()
belief = GaussianModelBelief(models=models)
agent = BayesDesigner(models=models, belief=belief, K=3, delta=0.05)
exp = PureExploreExperiment(agent=agent, environment=env, max_steps=100)
results = exp.run()
```

```python
# LP
env = GaussianEnvironment(means, std=1)
env.reset()
models = env.get_models()
belief = GaussianModelBelief(models=models)
lp_explorer = LPExplorer(models=models, belief=belief, K=3, latent_dim=3, delta=0.05)
lp_explorer.w_star, means = lp_explorer.solve_lp(None)
exp = PureExploreExperiment(agent=lp_explorer, environment=env, max_steps=100)
exp.run()
```

```python
N = 10 # runs per alg

ttts_log = []

bayes_log = []

kl_log = []

for i in range(N):
    env = GaussianEnvironment(means, std=1, seed=i)
    models = env.get_models()
    env.reset()
    
    # TTTS
    belief = GaussianModelBelief(models=models)
    ttts = TopTwoThompsonSampling(models=models, belief=belief, K=3, confidence=0.9)
    exp = PureExploreExperiment(agent=ttts, environment=env, max_steps=1000)
    log1 = exp.run()
    log1['Simple Regret'] = env.simple_regret(log1['Recommendation'])
    ttts_log.append(log1)
    
    
    env = GaussianEnvironment(means, std=1, seed=i)
    models = env.get_models()
    env.reset()
    # Bayes Design
    belief = GaussianModelBelief(models=models, seed=i)
    bayes = BayesDesigner(models=models, belief=belief, K=3, delta=0.1, n_draws=100)
    exp = PureExploreExperiment(agent=bayes, environment=env, max_steps=1000)
    log2 = exp.run()
    log2['Simple Regret'] = env.simple_regret(log2['Recommendation'])
    bayes_log.append(log2)
    
    env = GaussianEnvironment(means, std=1, seed=i)
    models = env.get_models()
    env.reset()
    # KL Exp
    belief = GaussianModelBelief(models=models, seed=i)
    kl_agent = DivergenceExplorer(models=models, belief=belief, K=3, confidence=0.9)
    exp = PureExploreExperiment(agent=kl_agent, environment=env, max_steps=1000)
    log3 = exp.run()
    log3['Simple Regret'] = env.simple_regret(log3['Recommendation'])
    kl_log.append(log3)
```

```python
import pandas as pd 
ttts_results = pd.DataFrame(ttts_log)
bayes_results = pd.DataFrame(bayes_log)
kl_results = pd.DataFrame(kl_log)
```

```python
import seaborn as sns
ttts_time = ttts_results['Stop time'].values
bayes_time = bayes_results['Stop time'].values
kl_time = kl_results['Stop time'].values
sns.distplot(ttts_time, label='TTTS')
sns.distplot(bayes_time, label='BayesDesigner')
sns.distplot(kl_time, label='Div Explorer')
print('Expected time DE: ', np.mean(kl_time))
print('Expected time TTTS: ', np.mean(ttts_time))
print('Expected time BD: ', np.mean(bayes_time))

plt.vlines(np.mean(ttts_time), 0, 1, linestyles='--', color='blue', label='Expected time TTTS')
plt.vlines(np.mean(bayes_time), 0, 1, linestyles='--', color='red', label='Expected time BD')
plt.vlines(np.mean(kl_time), 0, 1, linestyles='--', color='green', label='Expected time DivExp')


plt.xlabel('time')
plt.legend(bbox_to_anchor=(2.1, 1.1))
```

```python
ttts_correct = ttts_results['Correct Arm'].mean()
bayes_correct = bayes_results['Correct Arm'].mean()
kl_correct = kl_results['Correct Arm'].mean()
print(f'TTTS is correct: {ttts_correct} of the times')
print(f'BD is correct: {bayes_correct} of the times')
print(f'DivExp is correct: {kl_correct} of the times')
```
