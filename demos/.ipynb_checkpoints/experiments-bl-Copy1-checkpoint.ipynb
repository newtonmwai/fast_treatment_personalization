```python
N = 10000

e = ADCBEnvironment(
                    gamma=2, 
                    epsilon=0.1, 
                    policy='DX_Based',
                    regenerate=False,
                    horizon=6,
                    n_buffer=N,
                    sequential=False,
                    z_dim=6
                )
cs = e.reset()
gen_data = e.model_

#print('Time spent fitting env: %.2fs' %(e.fit_time))
P_Z = gen_data['Z'].value_counts(normalize=True)
P_Z = np.array(list(dict(sorted(P_Z.items())).values()))
print("P(Z): ", P_Z)

plt.figure(figsize = (8,6))
plt.hist(gen_data.loc[gen_data['VISCODE'] == 0][['Z']].values, bins=12)
plt.ylabel('')
plt.xlabel('Z')
plt.xlim([0, 5])

#plt.xticks(np.arange(0, 110, 10))
plt.grid(zorder=-100)
```
