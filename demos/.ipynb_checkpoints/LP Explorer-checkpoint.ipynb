```python
from healthy_gym.experiment import PureExploreExperimentADCB
from healthy_gym.agents.pure_explorer import *


Samples = 10# runs per alg
max_steps = 100
confidence=0.95
kl_log = []
ttts_log = []
lpexp_log = []

r_sigma =1 
env = ADCBEnvironment(
                    gamma=2, 
                    epsilon=0.1, 
                    policy='DX_Based',
                    regenerate=False,
                    horizon=6,
                    n_buffer=1000,
                    sequential=False,
                    reward_sigma=r_sigma,
                    z_dim=6)
for i in range(Samples):
    env.reset()
    Y_hat_model = env.load_cov_model('ADAS13', autoreg=True)
    models = env.get_models(env, Y_hat_model, r_sigma)
    
    # Divergence Explorer
    belief = GaussianModelBelief(models=models)
    kl_agent = DivergenceExplorer(models=models, belief=belief, confidence=confidence)
    exp1 = PureExploreExperimentADCB(agent=kl_agent, environment=env, max_steps=max_steps)
    log1 = exp1.run()
    #log1['Simple Regret'] = env.simple_regret(log1['Recommendation'])
    kl_log.append(log1)
    
    # Two-Top TS
    belief = GaussianModelBelief(models=models)
    ttts = TopTwoThompsonSampling(models=models, belief=belief, confidence=confidence)
    exp2 = PureExploreExperimentADCB(agent=ttts, environment=env, max_steps=max_steps)
    log2 = exp2.run()
    #log2['Simple Regret'] = env.simple_regret(log2['Recommendation'])
    ttts_log.append(log2)
    
    # LP Explorer
    belief = GaussianModelBelief(models=models)
    
    #belief = LatentModelBelief(models=models)
    lp_explorer = LPExplorer(models=models, belief=belief)
    # solve LP
    lp_explorer.w_star, means = lp_explorer.solve_lp(x)
    #belief.models = means
    
    exp3 = PureExploreExperimentADCB(agent=lp_explorer, environment=env, max_steps=max_steps)
    log3 = exp3.run()
    #log2['Simple Regret'] = env.simple_regret(log2['Recommendation'])
    lpexp_log.append(log3)
```

```python
import pandas as pd 
kl_results = pd.DataFrame(kl_log)
ttts_results = pd.DataFrame(ttts_log)
lpexp_results = pd.DataFrame(lpexp_log)

import seaborn as sns
kl_time = kl_results['Stop time'].values
ttts_time = ttts_results['Stop time'].values
lpexp_time = lpexp_results['Stop time'].values

plt.figure(figsize=(12,10))

sns.distplot(kl_time, label='Divergence Explorer')
sns.distplot(ttts_time, label='TTTS')
sns.distplot(lpexp_time, label='LP Explorer')
print('Expected time DE: ', np.mean(kl_time))
print('Expected time TTTS: ', np.mean(ttts_time))
print('Expected time LP: ', np.mean(lpexp_time))

plt.vlines(np.mean(kl_time), 0, 1, linestyles='--', color='blue', label='Expected time DE')
plt.vlines(np.mean(ttts_time), 0, 1, linestyles='--', color='red', label='Expected time TTTS')
plt.vlines(np.mean(lpexp_time), 0, 1, linestyles='--', color='k', label='Expected time LP')
#plt.xlim(0, 100)

plt.xlabel('time')
plt.legend(bbox_to_anchor=(1.55, 1.))
plt.grid(zorder=-100)
```

```python
kl_correct = kl_results['Correct Model'].mean()
ttts_correct = ttts_results['Correct Model'].mean()
lp_correct = lpexp_results['Correct Model'].mean()

print(f'Divergence Explorer is correct: {kl_correct} of the times')
print(f'TTTS is correct: {ttts_correct} of the times')
print(f'LP is correct: {lp_correct} of the times')
```
