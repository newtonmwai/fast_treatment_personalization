```python
%load_ext autoreload
%autoreload 2

import warnings
warnings.filterwarnings("ignore")

import os, sys, pickle
import seaborn as sns

sys.path.append('../')

from healthy_gym.agents.pure_explorer import *
from healthy_gym.agents.beliefs import *

from healthy_gym.environments.gaussian_environment import GaussianEnvironment

from healthy_gym.experiment import PureExploreExperiment

import matplotlib.pyplot as plt


%matplotlib inline

plt.rc('font', size=24, family='serif')
```

```python
means = np.array([[2.3, 2.5, 0, 1.5], [2.5, 2.3, 2, 1], [0, 0, 0.5, 2]])
plt.scatter([0, 1, 2, 3], means[0, :], label=r'$Z_0$') 
plt.scatter([0, 1, 2, 3], means[1, :], marker='*', label=r'$Z_1$')
plt.scatter([0, 1, 2, 3], means[2, :], marker='^', label=r'$Z_2$')
plt.ylabel(r'Expected reward')
plt.xlabel('Arm')
plt.legend(bbox_to_anchor=(1.1, 1.05))
```

```python
env = GaussianEnvironment(means, std=1)
env.reset()
models = env.get_models()
belief = GaussianModelBelief(models=models)
kl_agent = DivergenceExplorer(models=models, belief=belief, confidence=0.9)
exp = PureExploreExperiment(agent=kl_agent, environment=env, max_steps=100)
exp.run()
```

# Compare TTTS with DivergenceExplorer

```python
N = 1000 # runs per alg

kl_log = []

ttts_log = []

for i in range(N):
    env = GaussianEnvironment(means, std=1)
    models = env.get_models()
    env.reset()
    
    # Divergence Explorer
    belief = GaussianModelBelief(models=models)
    kl_agent = DivergenceExplorer(models=models, belief=belief, confidence=0.95)
    exp = PureExploreExperiment(agent=kl_agent, environment=env, max_steps=100)
    log1 = exp.run()
    log1['Simple Regret'] = env.simple_regret(log1['Recommendation'])
    kl_log.append(log1)
    
    # Two-Top TS
    belief = GaussianModelBelief(models=models)
    ttts = TopTwoThompsonSampling(models=models, belief=belief, confidence=0.95)
    exp = PureExploreExperiment(agent=ttts, environment=env, max_steps=100)
    log2 = exp.run()
    log2['Simple Regret'] = env.simple_regret(log2['Recommendation'])
    ttts_log.append(log2)
```

```python
import pandas as pd 
kl_results = pd.DataFrame(kl_log)
ttts_results = pd.DataFrame(ttts_log)
```

# Distribution of stopping time

```python
import seaborn as sns
kl_time = kl_results['Stop time'].values
ttts_time = ttts_results['Stop time'].values
sns.distplot(kl_time, label='Divergence Explorer')
sns.distplot(ttts_time, label='TTTS')
print('Expected time DE: ', np.mean(kl_time))
print('Expected time TTTS: ', np.mean(ttts_time))

plt.vlines(np.mean(kl_time), 0, 1, linestyles='--', color='blue', label='Expected time DE')
plt.vlines(np.mean(ttts_time), 0, 1, linestyles='--', color='red', label='Expected time TTTS')
plt.xlim(0, 100)

plt.xlabel('time')
plt.legend(bbox_to_anchor=(2.1, 1.1))
```

# Distribution of Correct

```python
kl_correct = kl_results['Correct Model'].mean()
ttts_correct = ttts_results['Correct Model'].mean()
print(f'Divergence Explorer is correct: {kl_correct} of the times')
print(f'TTTS is correct: {ttts_correct} of the times')
```

```python
from healthy_gym.environments.adcb import *
sample_size = 1000
e = ADCBEnvironment(gamma=2, 
                    epsilon=0.1, 
                    policy='DX_Based',
                    regenerate=False,
                    horizon=6,
                    n_buffer=sample_size,
                    rnd_seed=1)
model = e.load_cov_model('ADAS13', autoreg=True)
print('Time spent fitting env: %.2fs' %(e.fit_time))
z_space = get_models(e, model, 2)
x = e.reset()
```

```python
N = 5


kl_log = []
ttts_log = []

for i in range(N):
    env = GaussianEnvironment(means, std=1)
    models = env.get_models()
    env.reset()
    
    # Divergence Explorer
    belief = GaussianModelBelief(z_space, std=1)
    kl_agent = DivergenceExplorer(models=models, belief=belief, confidence=0.9)
    exp = PureExploreExperiment(agent=kl_agent, environment=env, max_steps=100)
    log1 = exp.run()
    kl_log.append(log1)
    
    # Two-Top TS
    belief = GaussianModelBelief(z_space, std=1)
    ttts = TopTwoThompsonSampling(models=models, belief=belief, confidence=0.9)
    exp = PureExploreExperiment(agent=ttts, environment=env, max_steps=100)
    log2 = exp.run()
    ttts_log.append(log2)
```
