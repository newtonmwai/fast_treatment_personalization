```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split

Categorical_cols = {'PTGENDER': 2, 'PTEDUCAT': 13, 'PTETHCAT': 3, 'PTRACCAT': 7, 'PTMARRY': 5, 'DX': 3, 'PTGENDER_prev': 2,
                    'PTEDUCAT_prev': 13, 'PTETHCAT_prev': 3, 'PTRACCAT_prev': 7, 'PTMARRY_prev': 5, 'APOE4_prev': 3, 'prev_DX': 3, 'A_Cat_prev': 8}

#grouped_id_data = data.groupby('RID')
#datanew = pd.DataFrame({})
#for i in range(sample_size):
#    df = grouped_id_data.get_group(i)
#    df['prev_DX'] = df['DX'].shift(periods=1).fillna(-1)
#    df['prev_Y_hat'] = df['Y_hat'].shift(periods=1).fillna(170)
#    datanew = datanew.append(df, ignore_index=True)
    
def check_categorical(df, cols, Y_col, categorical_cols=Categorical_cols):
    for col in cols:
        #print("\ncheck_categorical: ", col)
        # OR any(s in col for s in categorical_cols)
        if(((col in categorical_cols)) and (col != Y_col)):
            #print("\ncheck_categorical,  True: ", col)
            df = df.astype({col: int})
            # print(df.head())
            df = pd.concat([df, pd.get_dummies(df[col].astype(pd.CategoricalDtype(
                categories=list(range(categorical_cols[col])))), prefix=col, drop_first=True)], axis=1)
            #df = pd.concat([df, pd.get_dummies(df[col], prefix=col, drop_first=True)], axis=1)
            df.drop([col], axis=1, inplace=True)
    return df

def pre_processing(df1):
    # Columns to be standardized
    standardize_x_cols = ['TAU', 'PTAU', 'FDG', 'AV45', 'prev_Y_hat']

    df = df1.copy()
    
    sc = StandardScaler()
    df[standardize_x_cols] = sc.fit_transform(df[standardize_x_cols])
    # creating instance of labelencoder
    labelencoder = LabelEncoder()
    df['A_Cat'] = labelencoder.fit_transform(df['A'])
    df.drop('A', axis=1, inplace=True)
    
    df = check_categorical(df, X_cols, 'Y_hat')    
    return df
```

# Linear Contextual TS & mTS

### Helper methods for the experiments

```python
def linearContextualBanditExperiment(x, dimension, X_cols, T, delta, z, trueZ=False):
    dimension=dimension
    regrets = np.zeros(T)
    rewards = np.zeros(T)
    expected_rewards = np.zeros(T+1)
    expected_regrets = np.zeros(T+1)
    
    x = x[X_cols]
    if trueZ:
        dimension+=1
        x['Z'] = z
    
    x = np.array(dm.check_categorical(x, list(x.columns), '_'))
    x = x.reshape(dimension,)
    
    beliefs = [LinearGaussianBelief(d=dimension) for i in range(A)]
    context_ts = ThompsonSampling(beliefs=beliefs)

    for t in range(T):
        action, _, info = context_ts.act(x)
        _, r_env, _, info = e.step(action)
        
        r = delta[action]
        regret = np.max(delta) - r
        
        r_avg = 0
        regret_avg = 0
        
        if(t>0):
            r_avg = expected_rewards[-1] + \
                (r - expected_rewards[-1]) / (t + 1)

            regret_avg = expected_regrets[-1] + \
                (regret - expected_regrets[-1]) / (t + 1)        
        
        context_ts.update(x, action, r_env) 
        
        rewards[t] = r
        regrets[t] = regret
        expected_rewards[t+1] = r_avg
        expected_regrets[t+1] = regret_avg
        
    return regrets, rewards, expected_regrets, expected_rewards

def mTSExperiment(x, z, model, T, delta, sigma, prior=None, trueZ=False):
    regrets = np.zeros(T)
    rewards = np.zeros(T)
    
    expected_rewards = np.zeros(T+1)
    expected_regrets = np.zeros(T+1)
        
    belief = LatentModelBelief(model, random_state=1, sigma=sigma, prior=prior) # , 
    model_ts = ModelThompsonSampling2(model=model, belief=belief)

    for t in range(T):
        if(trueZ):
            action, _, info_act = model_ts.act(x, z=z)
        else:
            action, _, info_act = model_ts.act(x)
            
        _, r_env, _, info = e.step(action)
        
        r = delta[action]
        regret = np.max(delta) - r
        
        r_avg = 0
        regret_avg = 0
        
        if(t>0):
            r_avg = expected_rewards[-1] + \
                (r - expected_rewards[-1]) / (t + 1)

            regret_avg = expected_regrets[-1] + \
                (regret - expected_regrets[-1]) / (t + 1)
        
        model_ts.update(x, info_act['z'], action, r_env)
        
        rewards[t] = r
        regrets[t] = regret
        expected_rewards[t+1] = r_avg
        expected_regrets[t+1] = regret_avg
        
    return regrets, rewards, expected_regrets, expected_rewards

def mean_std_df(x):
    df = pd.DataFrame(x)
    df['mean'] = df.mean(axis=1)
    if(df.count(axis='columns')[0])>2:
        df['std'] = df[map(lambda x :x not in ['mean'], list(df.columns))].std(axis=1)
    return df
```

#### Running mTS and Linear Contextual Bandit
- $\sigma_{reward}$ is the environment's $\sigma$
- $\sigma$ is the $\sigma$ used to compute the likelihood 
- $true_z$ is a boolean to switch between Actual Z ot learning a Z
- $N_{samples}$ is the number of samples over which mean reward & regret are computed
- T is the number of trials

```python
reward_sigmas = [1, 2, 3, 4]
df_results = pd.DataFrame({})
N = 10000

for reward_sigma in reward_sigmas:
    e = ADCBEnvironment(gamma=2, 
                    epsilon=0.1, 
                    policy='DX_Based',
                    regenerate=False,
                    horizon=6,
                    n_buffer=N,
                    sequential=False,
                    z_dim=6
                )

    print('Time spent fitting env: %.2fs' %(e.fit_time))
    cs = e.reset()

    N_samples = 10
    T = 100
    A = 8
    dimension = 30

    rfr = e.load_cov_model('ADAS13', autoreg=True)

    N_rewards_lcb = {}
    N_regrets_lcb = {}
    N_rewards_mTS = {}
    N_regrets_mTS = {}

    N_expected_rewards_lcb = {}
    N_expected_regrets_lcb = {}
    N_expected_rewards_mTS = {}
    N_expected_regrets_mTS = {}

    sigma = reward_sigma
    prior = None #[0.57, 0.43]
    truez = False
    for p in range(N_samples):
        cs = e.reset()
        x = e.buffer_
        z = x['Z'].values

        model = SimpleModel2(e, rfr)
        delta = model.predict(x, z)

        regrets_mTS, rewards_mTS, expected_regrets_mTS, expected_rewards_mTS = mTSExperiment(x, z, model, T, delta, sigma, prior=prior, trueZ=truez)
        regrets_lcb, rewards_lcb, expected_regrets_lcb, expected_rewards_lcb = linearContextualBanditExperiment(x, dimension, X_cols, T, delta, z, trueZ=truez)

        N_rewards_mTS[p] = rewards_mTS
        N_regrets_mTS[p] = regrets_mTS
        N_rewards_lcb[p] = rewards_lcb
        N_regrets_lcb[p] = regrets_lcb

        N_expected_rewards_mTS[p] = expected_rewards_mTS
        N_expected_regrets_mTS[p] = expected_regrets_mTS
        N_expected_rewards_lcb[p] = expected_rewards_lcb
        N_expected_regrets_lcb[p] = expected_regrets_lcb

    rewards_mTS, regrets_mTS, expected_rewards_mTS, expected_regrets_mTS = [mean_std_df(x) for x in [N_rewards_mTS, N_regrets_mTS, N_expected_rewards_mTS, N_expected_regrets_mTS]]
    rewards_lcb, regrets_lcb, expected_rewards_lcb, expected_regrets_lcb = [mean_std_df(x) for x in [N_rewards_lcb, N_regrets_lcb, N_expected_rewards_lcb, N_expected_regrets_lcb]]

    df_results['rewards_mTS_reward_sigma_'+str(reward_sigma)+'_N_'+str(N_samples)] =  rewards_mTS['mean'].values
    df_results['regrets_mTS_reward_sigma_'+str(reward_sigma)+'_N_'+str(N_samples)] =  regrets_mTS['mean'].values
    if reward_sigma==1:
        df_results['rewards_lcb_'+str(N_samples)] =  rewards_lcb['mean'].values
        df_results['regrets_lcb_'+str(N_samples)] =  regrets_lcb['mean'].values
```
