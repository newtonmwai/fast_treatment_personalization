```python
%load_ext autoreload
%autoreload 2

import os, sys, pickle
import seaborn as sns

sys.path.append('../')

from healthy_gym.agents.thompson_sampling import *
from healthy_gym.agents.beliefs import *

import matplotlib.pyplot as plt

from healthy_gym.environments.adcb import *

%matplotlib inline

plt.rc('font', size=24, family='serif')
plt.rcParams["figure.figsize"] =(12, 8)
plt.style.use('tableau-colorblind10')

from healthy_gym.experiment import RegretMinExperiment
```

We fit the model

```python
sample_size = 1000
e = ADCBEnvironment(gamma=2, 
                    epsilon=0.1, 
                    policy='DX_Based',
                    regenerate=False,
                    horizon=6,
                    n_buffer=sample_size,
                    rnd_seed=1)
model = e.load_cov_model('ADAS13', autoreg=True)
print('Time spent fitting env: %.2fs' %(e.fit_time))
```

## Create the latent space

```python
z_space = get_models(e, model, 2)
x = e.reset()
#z_space[0].predict(x)
```

```python
model_belief = GaussianModelBelief(z_space, std=1)
model_thompson = ModelThompsonSampling(z_space, model_belief)
# Test 
action,_ ,_ = model_thompson.act(x)
```

Run Model-based experiment

```python
exp = RegretMinExperiment(model_thompson, e, n_steps=10)
model_results = exp.run()
```

```python
import matplotlib.pyplot as plt
plt.plot(model_results['cumulative_regret'], label='mTS')
plt.xlim(0, 10)
plt.legend()
```

```python
import seaborn as sns
sns.histplot(model_results['action'])
plt.title('Action distribution')
```
